```python
import pandas as pd
```

```python
url = 'http://www.espn.com/nba/attendance/_/year/2020' #set the url 
```

```python
html = pd.read_html(url) #use read html function 
html 
```

```python
#set columns name
columns = ['RK','Home_GMS', 'Home_total', 'Home_AVG', 'Home_PCT', 
            'Road_GMS', 'Road_total', 'Road_AVG', 'Road_PCT',
            'Overall_GMS','Overall_AVG', 'Overall_PCT']
```

```python
df1 = html[0]
df1
```

```python
df1.columns = columns #replace columns name
```

```python
final_df = df1[2:] #remove first two rows, only keep data that we want
final_df
```

```python
final_df.to_csv('2020_NBA_Attendance', sep='\t', encoding='utf-8')
```
