```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.espn.com/mlb/history/leaders/_/breakdown/season/year/2020/start/1' #set the url
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
players = soup.find_all('tr', attrs = {'class' : re.compile('row player-10-')})
players
```

```python
header = soup.find('tr', attrs={'class' : 'colhead'})
columns = [col.get_text() for col in header.find_all('td')]
df = pd.DataFrame(columns = columns)
print(df)
```

```python
#get each player row
for player in players:
    players_df = [data.get_text() for data in player.find_all('td')]
    df1 = pd.DataFrame(players_df).transpose()
    df1.columns = columns
    
    df = pd.concat([df, df1], ignore_index = True)
df
```

```python
df.to_csv('2020_NFL_players', sep='\t', encoding='utf-8')
```
